```python
import pandas as pd
df = pd.read_csv("../data/processed/online_retail_clean.csv")
df["invoicedate"] = pd.to_datetime(df["invoicedate"])
df.head()
total_revenue = df["total_sales"].sum()
total_orders = df["invoiceno"].nunique()
total_customers = df["customerid"].nunique()
avg_order_value = total_revenue / total_orders

total_revenue, total_orders, total_customers, avg_order_value
df["order_month"] = df["invoicedate"].dt.to_period("M").astype(str)

rev_by_month = (
    df.groupby("order_month", as_index=False)["total_sales"]
      .sum()
      .sort_values("total_sales", ascending=False)
      .head(6)
)
rev_by_month
top_countries = (
    df.groupby("country", as_index=False)["total_sales"]
      .sum()
      .sort_values("total_sales", ascending=False)
      .head(10)
)
top_countries
orders_per_customer = df.groupby("customerid")["invoiceno"].nunique()
returning_customers_pct = (orders_per_customer.gt(1).mean()) * 100
returning_customers_pct
insights = [
    f"Total revenue: {total_revenue:,.2f}",
    f"Total orders: {total_orders:,}",
    f"Total customers: {total_customers:,}",
    f"Average order value: {avg_order_value:,.2f}",
    f"Returning customers: {returning_customers_pct:.1f}%",
    f"Top country by revenue: {top_countries.iloc[0]['country']}",
    f"Best month by revenue: {rev_by_month.iloc[0]['order_month']}",
]
insights
```
