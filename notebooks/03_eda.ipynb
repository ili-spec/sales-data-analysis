```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("../data/processed/online_retail_clean.csv")
df.head()
monthly_sales = (
    df.groupby("order_month", as_index=False)["total_sales"].sum()
)

monthly_sales.head()
plt.figure(figsize=(10,5))
plt.plot(monthly_sales["order_month"], monthly_sales["total_sales"])
plt.xticks(rotation=45)
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.tight_layout()
plt.show()
```

```python
top_products = (
    df.groupby("description", as_index=False)["total_sales"]
      .sum()
      .sort_values("total_sales", ascending=False)
      .head(10)
)

top_products
plt.figure(figsize=(10,5))
plt.barh(top_products["description"], top_products["total_sales"])
plt.gca().invert_yaxis()
plt.title("Top 10 Products by Sales")
plt.xlabel("Total Sales")
plt.tight_layout()
plt.show()
top_customers = (
    df.groupby("customerid", as_index=False)["total_sales"]
      .sum()
      .sort_values("total_sales", ascending=False)
      .head(10)
)

top_customers
df["total_sales"].describe()


```
