```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("../data/raw/online_retail.xlsx.xlsx")
df.head()

```

```python
df.columns = [c.strip().lower().replace(" ","_") for c in df.columns]
df.columns
```

```python
missing = df.isna().sum().sort_values(ascending= False)
missing.head(10),df.duplicated().sum()
```

```python
df=df.dropna(subset = ["customerid", "invoiceno", "stockcode", "quantity", "unitprice", "invoicedate"])
df["costumerid"]=df["customerid"].astype("int64")
df = df[(df["quantity"] > 0) & (df["unitprice"] > 0)]
df.shape
```

```python
df["invoicedate"] = pd.to_datetime(df["invoicedate"])
df["order_date"] = df["invoicedate"].dt.date
df["order_month"] = df["invoicedate"].dt.to_period("M").astype("str")
df["order_month"] = df["invoicedate"].dt.year
df["order_weekday"] = df["invoicedate"].dt.day_name()
df.head
```

```python
df["total_sales"] = df["unitprice"] * df["quantity"]
df[["quantity","unitprice","total_sales"]].head()
```

```python
out_path =  "../data/processed/online_retail_clean.csv"
df.to_csv(out_path,index = False)
out_path
```
